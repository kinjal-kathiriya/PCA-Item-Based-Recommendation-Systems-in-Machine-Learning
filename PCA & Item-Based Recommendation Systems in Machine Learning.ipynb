```python
def euclidSim(inA, inB):
    return 1.0 / (1.0 + la.norm(inA - inB))

def pearsonSim(inA, inB):
    if len(inA) < 3: return 1.0
    return 0.5 + 0.5 * np.corrcoef(inA, inB, rowvar=0)[0][1]

def cosineSim(inA, inB):
    num = float(inA.T * inB)
    denom = la.norm(inA) * la.norm(inB)
    return 0.5 + 0.5 * (num / denom)
```

```python
def standEst(dataMat, user, simMeas, item):
    n = np.shape(dataMat)[1]
    simTotal = 0.0; ratSimTotal = 0.0
    data = np.matrix(dataMat)  # Use np.matrix explicitly
    for j in range(n):
        userRating = data[user,j]
        if userRating == 0: continue
        overLap = np.nonzero(np.logical_and(data[:,item]>0, data[:,j]>0))[0]
        if len(overLap) == 0:
            similarity = 0
        else:
            similarity = simMeas(data[overLap,item], data[overLap,j])
        simTotal += similarity
        ratSimTotal += similarity * userRating
    if simTotal == 0: return 0
    else: return ratSimTotal/simTotal
```

```python
def recommend(dataMat, user, N=3, simMeas=pearsonSim, estMethod=standEst):
    unratedItems = np.nonzero(dataMat[user,:].A==0)[1]
    if len(unratedItems) == 0: return 'you rated everything'
    itemScores = []
    for item in unratedItems:
        estimatedScore = estMethod(dataMat, user, simMeas, item)
        itemScores.append((item, estimatedScore))
    return sorted(itemScores, key=lambda jj: jj[1], reverse=True)[:N]
```

```python
def svdEst(dataMat, user, simMeas, item):
    # Perform SVD on the data matrix
    U, Sigma, VT = la.svd(dataMat)
    
    # Keep only the top 4 singular values (arbitrary choice for dimensionality reduction)
    k = 4
    Sigma_k = np.diag(Sigma[:k])
    U_k = U[:, :k]
    VT_k = VT[:k, :]
    
    # Reconstruct the reduced matrix
    dataMat_reduced = np.dot(np.dot(U_k, Sigma_k), VT_k)
    
    # Convert to matrix for consistency
    data = np.matrix(dataMat_reduced)
    
    # Compute similarity and prediction as in standEst
    n = np.shape(data)[1]
    simTotal = 0.0; ratSimTotal = 0.0
    for j in range(n):
        userRating = data[user, j]
        if userRating == 0: continue
        overLap = np.nonzero(np.logical_and(data[:, item] > 0, data[:, j] > 0))[0]
        if len(overLap) == 0:
            similarity = 0
        else:
            similarity = simMeas(data[overLap, item], data[overLap, j])
        simTotal += similarity
        ratSimTotal += similarity * userRating
    if simTotal == 0: return 0
    else: return ratSimTotal / simTotal
```

```python
def test(dataMat, estMethod, simMeas):
    """
    Iterates over all users, performs cross-validation for each using cross_validate_user,
    and returns error information for computing MAE.
    
    Parameters:
    - dataMat: Rating matrix (numpy matrix, shape: users x items)
    - estMethod: Estimation method (e.g., standEst or svdEst)
    - simMeas: Similarity measure (e.g., pearsonSim)
    
    Returns:
    - total_error: Sum of absolute errors across all predictions
    - total_predictions: Total number of predictions made
    """
    total_error = 0.0
    total_predictions = 0
    
    # Iterate over all users (0 to 999 for 1000 users)
    for user in range(dataMat.shape[0]):
        # Call cross_validate_user for the current user
        error, num_preds = cross_validate_user(dataMat, user, estMethod, simMeas)
        
        # Aggregate error and prediction counts
        total_error += error
        total_predictions += num_preds
    
    return total_error, total_predictions
```

```python
def cross_validate_user(dataMat, user, estMethod, simMeas):
    """
    Performs leave-one-out cross-validation for a single user and returns error metrics.
    
    Parameters:
    - dataMat: Rating matrix
    - user: User ID (integer)
    - estMethod: Estimation method
    - simMeas: Similarity measure
    
    Returns:
    - total_error: Sum of absolute errors for predictions
    - num_preds: Number of predictions made
    """
    total_error = 0.0
    num_preds = 0
    
    # Get items rated by the user
    rated_items = np.nonzero(dataMat[user, :].A != 0)[1]
    
    for item in rated_items:
        # Hold out the current rating
        actual_rating = dataMat[user, item]
        dataMat[user, item] = 0  # Temporarily set to unrated
        
        # Predict the rating
        predicted_rating = estMethod(dataMat, user, simMeas, item)
        
        # Calculate absolute error
        total_error += abs(predicted_rating - actual_rating)
        num_preds += 1
        
        # Restore the original rating
        dataMat[user, item] = actual_rating
    
    return total_error, num_preds
```

```python

def print_most_similar_jokes(dataMat, jokes, query_joke_id, k, simMeas):
    """
    Outputs the query joke and the top k most similar jokes based on user ratings.
    
    Parameters:
    - dataMat: Rating matrix (numpy matrix, shape: users x items)
    - jokes: Array of joke texts (length: number of items)
    - query_joke_id: ID of the query joke (integer, 0-99)
    - k: Number of similar jokes to output (integer)
    - simMeas: Similarity metric function (e.g., pearsonSim)
    """
    # Validate inputs
    if not 0 <= query_joke_id < dataMat.shape[1]:
        print(f"Error: Joke ID {query_joke_id} is out of range (0-{dataMat.shape[1]-1})")
        return
    if k <= 0 or k > dataMat.shape[1] - 1:
        print(f"Error: k ({k}) must be between 1 and {dataMat.shape[1]-1}")
        return
    
    # Get ratings for the query joke
    query_ratings = dataMat[:, query_joke_id]
    
    # Compute similarity with all other jokes
    similarities = []
    for joke_id in range(dataMat.shape[1]):
        if joke_id == query_joke_id:
            continue  # Skip the query joke itself
        other_ratings = dataMat[:, joke_id]
        similarity = simMeas(query_ratings, other_ratings)
        similarities.append((joke_id, similarity))
    
    # Sort by similarity in descending order
    similarities.sort(key=lambda x: x[1], reverse=True)
    
    # Print the query joke
    print(f"\nQuery Joke (ID: {query_joke_id}):")
    print(jokes[query_joke_id])
    print("-" * 50)
    
    # Print the top k most similar jokes
    print(f"\nTop {k} Most Similar Jokes:")
    for i, (joke_id, sim_score) in enumerate(similarities[:k], 1):
        print(f"\n{i}. Joke ID: {joke_id}, Similarity: {sim_score:.4f}")
        print(jokes[joke_id])
        print("-" * 50)
```

```python
def train_item_similarities(dataMat, simMeas):
    n_items = dataMat.shape[1]
    sim_matrix = np.zeros((n_items, n_items))
    for i in range(n_items):
        for j in range(i, n_items):
            similarity = simMeas(dataMat[:, i], dataMat[:, j])
            sim_matrix[i, j] = similarity
            sim_matrix[j, i] = similarity
    return sim_matrix

def item_based_est(dataMat, user, item, k, sim_matrix):
    if not 0 <= item < dataMat.shape[1] or not 0 <= user < dataMat.shape[0]:
        print(f"Invalid user {user} or item {item}")
        return 0
    
    user_ratings = dataMat[user, :].A[0]
    sim_scores = sim_matrix[item, :]
    
    similar_items = [(i, float(sim_scores[i])) for i in range(len(sim_scores)) if i != item and user_ratings[i] != 0]
    
    if not similar_items:
        print(f"User {user}, Item {item}: No similar items found")
        return 0
    
    similar_items.sort(key=lambda x: x[1], reverse=True)
    num_to_take = builtins.min(k, len(similar_items))  # Use Python's min
    top_k = similar_items[:num_to_take]
    
    sim_total = 0.0
    weighted_rating = 0.0
    for sim_item, sim_score in top_k:
        rating = user_ratings[sim_item]
        sim_total += sim_score
        weighted_rating += sim_score * rating
    
    return weighted_rating / sim_total if sim_total > 0 else 0
```

```python
# Function to compute item-item similarity and store it in a dictionary
def compute_item_similarity(dataMat, sim_function):
    """
    Computes item-item similarity matrix using the specified similarity function.
    """
    num_items = dataMat.shape[1]
    similarity_matrix = np.zeros((num_items, num_items))
    
    for i in range(num_items):
        for j in range(i + 1, num_items):
            mask = (dataMat[:, i] > 0) & (dataMat[:, j] > 0)  # Users who rated both items
            if np.any(mask):
                similarity_matrix[i, j] = sim_function(dataMat[mask, i], dataMat[mask, j])
                similarity_matrix[j, i] = similarity_matrix[i, j]  # Symmetric matrix
    
    np.fill_diagonal(similarity_matrix, 1.0)  # Self-similarity is 1
    return similarity_matrix
```

```python
# Function to estimate item-based rating
def item_based_est(dataMat, user, item, k, sim_matrix):
    """
    Estimates rating for a specific item for a given user using item-based collaborative filtering.
    """
    if item >= dataMat.shape[1]:  # Ensure item index is valid
        return 0
    
    user_ratings = dataMat[user]
    similar_items = np.argsort(sim_matrix[item])[-(k + 1):-1]  # Top k similar items
    
    numerator = np.sum(sim_matrix[item, similar_items] * user_ratings[similar_items])
    denominator = np.sum(np.abs(sim_matrix[item, similar_items]))
    
    return numerator / denominator if denominator != 0 else 0
```

```python
def predict_ratings(user_ratings, similarity_matrix, top_k=5):
    """
    Predicts user ratings using item-based collaborative filtering.
    """
    num_items = similarity_matrix.shape[0]
    predicted_ratings = np.zeros(num_items)
    
    for item in range(num_items):
        if user_ratings[item] == 0:  # Only predict for unrated items
            sim_scores = similarity_matrix[item]
            rated_items = user_ratings > 0
            top_similar_items = np.argsort(sim_scores[rated_items])[-top_k:]  # Top K similar items
            
            if len(top_similar_items) > 0:
                weighted_sum = np.sum(sim_scores[rated_items][top_similar_items] * user_ratings[rated_items][top_similar_items])
                sum_weights = np.sum(np.abs(sim_scores[rated_items][top_similar_items]))
                predicted_ratings[item] = weighted_sum / sum_weights if sum_weights != 0 else 0
    
    return predicted_ratings
```

```python

# Similarity Functions
def cosine_similarity(inA, inB):
    num = np.dot(inA, inB)
    denom = np.linalg.norm(inA) * np.linalg.norm(inB)
    return num / denom if denom != 0 else 0

def pearson_similarity(inA, inB):
    if len(inA) < 3: return 1.0
    return np.corrcoef(inA, inB)[0, 1]
```

```python
def cross_validate_user(dataMat, user, estMethod, sim_matrix, k, test_ratio=0.2):
    total_error = 0.0
    num_preds = 0
    rated_items = np.nonzero(dataMat[user, :] != 0)[0]  # Fix: Remove .A
    if len(rated_items) == 0:
        return 0.0, 0
    
    np.random.shuffle(rated_items)
    test_size = int(len(rated_items) * test_ratio)
    test_items = rated_items[:test_size]
    dataMat_copy = dataMat.copy()
    for item in test_items:
        dataMat_copy[user, item] = 0
    
    for item in test_items:
        actual_rating = dataMat[user, item]
        predicted_rating = estMethod(dataMat_copy, user, item, k, sim_matrix)
        total_error += abs(predicted_rating - actual_rating)
        num_preds += 1
    
    return total_error, num_preds

# Testing function
def test(dataMat, estMethod, sim_matrix, k):
    total_error = 0.0
    total_predictions = 0
    for user in range(dataMat.shape[0]):
        error, num_preds = cross_validate_user(dataMat, user, estMethod, sim_matrix, k, test_ratio=0.2)
        total_error += error
        total_predictions += num_preds
    return total_error, total_predictions
```

```python
# Evaluate model using MAE for different similarity measures
for sim_name, sim_matrix in similarity_matrices.items():
    total_error, total_preds = test(dataMat, item_based_est, sim_matrix, k=10)
    mae = total_error / total_preds if total_preds > 0 else 0
    print(f"{sim_name} MAE (k=10): {mae:.4f}")
```

```python
def recommend(dataMat, user, sim_matrix, k, N=3, estMethod=item_based_est):
    """
    Recommend top-N items for a user based on predicted ratings.
    """
    unrated_items = np.nonzero(dataMat[user, :] == 0)[0]  # Fix: Remove .A
    if len(unrated_items) == 0:
        return "You rated everything"
    
    item_scores = []
    for item in unrated_items:
        predicted_score = estMethod(dataMat, user, item, k, sim_matrix)
        item_scores.append((item, predicted_score))
    
    return sorted(item_scores, key=lambda x: x[1], reverse=True)[:N]
```
